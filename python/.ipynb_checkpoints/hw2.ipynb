```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import random

# fine the connection components in the image
def connected_components(original_image):
    # list of tags we have used 
    tags = []
    # current  tag (remember 1 and 0 are already in image so start from 2)
    tag = 2
    # counter
    cntr = 0
    # convert image to numpy array
    imageArr = np.array(original_image)

    # Define the amount of padding for each side (top, bottom, left, right)
    top_padding = 1
    bottom_padding = 1
    left_padding = 1
    right_padding = 1

    # Pad 'image1' with zeros
    image = np.pad(imageArr, ((top_padding, bottom_padding), (left_padding, right_padding)), mode='constant')

    print("image Shape: ", image.shape[0])
    print("image:", image[0,0])
    print("image:", image[1,1])
    print("image:", image[2,2])
    print("image:", image[3,3])

#   Loop through the image numpy array 
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            #simple threshold if pixel value is less than 96 then set it to zero otherwise set it to be 1
            if image[i, j] < 96:
                image[i, j] = 0
            else :
                image[i, j] = 1

            if image[i, j] != 0:
                # if the pixel to the left and the pixel above the current pixel are not equal to zero then if the 2 pixel are different
                # then set the current pixel equal to the one above it
                if i != 0 and j != 0 and image[i, j-1] != 0 and image[i-1, j] != 0 and image[i-1, j] != image[i, j-1]:
                    image[i, j] = image[i, j - 1]
                    # remove the pixel to the left
                    tags.remove(image[i - 1, j])
                    # reduce the counter value
                    cntr -= 1
                    # go thru the image and set all references of the left pixels tag number to the value of the pixel above
                    image[image == image[i - 1, j]] = image[i, j]
                # if the pixel to the left and the pixel above the current pixel are not equal to zero then if the 2 pixel are different
                # then set the current pixel equal to the one above it
                if i != 0 and j != 0 and image[i+1, j-1] != 0 and image[i-1, j] != 0 and image[i-1, j] != image[i+1, j-1]:
                    image[i, j] = image[i+1, j - 1]
                    # remove the tag for pixel to the left
                    tags.remove(image[i - 1, j])
                    # reduce the counter value, since a new pontential connnected component has been confirmed not to exist and is actaully part of an old one
                    cntr -= 1
                    # go thru the image and set all references of the left pixels tag number to the value of the pixel above
                    image[image == image[i - 1, j]] = image[i, j]
                elif i != 0 and image[i-1, j] != 0:
                    image[i, j] = image[i-1, j]
                elif j != 0 and image[i, j-1] != 0:
                    image[i, j] = image[i, j-1]
                # the pixel that is above and to the left
                elif j != 0 and image[i-1, j-1] != 0:
                    image[i, j] = image[i-1, j-1]
                elif j != 0 and image[i+1, j-1] != 0:
                    image[i, j] = image[i+1, j-1]
                else:
                    image[i, j] = tag
                    tags.append(tag)
                    tag += 1
                    cntr += 1
    return image, tags, cntr

# Path to the binary file
file_path = 'Suzi1.bin'  # Replace with the correct path

# Read the binary file
with open(file_path, 'rb') as file:
    data = np.fromfile(file, dtype=np.uint8)

# Reshape data into a 256x256 grayscale images
original_image = data.reshape((256, 256))

#  find the the connected components in the original image
result1 = connected_components(original_image)

#store the the result of the connected_components function for the original image
new_array, tags, cntr = result1 

print("Tags:", tags)
print("Counter:", cntr)

# Create a color map for the components
component_colors = {}
for tag in tags:
    component_colors[tag] = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))

# Create a new image with colored components
colored_image = Image.new('RGB', (new_array.shape[1], new_array.shape[0]))

for y in range(new_array.shape[0]):
    for x in range(new_array.shape[1]):
        tag = new_array[y, x]
        if tag != 0:
            colored_image.putpixel((x, y), component_colors[tag])

# Display the original and new images
# Display the image
plt.imshow(original_image, cmap='gray')
plt.title('Suzi Grayscale Image')
plt.show()

# original_image.show(title="Original Image")
colored_image.show(title="Colored Components Image")

# Save the new image to a file (optional)
colored_image.save("output_colored_image.bmp")
```
